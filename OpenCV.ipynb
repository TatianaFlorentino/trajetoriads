```
# importar pacotes necessários
import numpy as np
import matplotlib.pyplot as plt

# criar um set entre -10 e 10 e aplicar a função sigmóide
x = np.arange(-10, 10)
y = 1 / (1 + np.exp(-x))

# plotar a curva sigmoidal
plt.plot(x, y)
plt.vlines(0, 0, 1, colors='r', linestyles='dashed')
plt.show()
```
